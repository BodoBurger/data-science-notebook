# Retrieve shapes from OpenStreetMap by tags

```python
import geopandas as gpp
import osmnx as ox
```

```python
place = "Freiburg i. Br., Baden-WÃ¼rttemberg, Germany"
place_gdf = ox.geocode_to_gdf(place)   
```

```python
place_poly = place_gdf.geometry.iloc[0]
```

```python
tags = {"admin_level": "10"}
```

```python
gdf = ox.features_from_polygon(place_poly, tags)
```

```python
gdf.explore(tooltip=["name"])
```
